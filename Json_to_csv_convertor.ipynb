```python
def inning_df(file):
    return pd.DataFrame(file)
```

```python
def over_df(file):
    return pd.DataFrame(file)
```

```python
def player_out(x):
    try:
        if math.isnan(x):
            return 0
    except TypeError:     
        return x[0].get('player_out', 0)
    
def wicket_type(x):
    try:
        if math.isnan(x):
            return 0
    except TypeError:     
        return x[0].get('kind', 0)
    
def fielders(x):
    try:
        if math.isnan(x):
            return 0
    except TypeError:     
        result = x[0].get('fielders', 0)
        if result == 0:
            return 0
        else:
            temp = ''
            for i in result:
                temp = temp + i['name'] + ', '

            return temp[:-2]
```

```python
def wides(x):
    try:
        if math.isnan(x):
            return 0
    except TypeError:     
        return x.get('wides', 0)
    
def no_ball(x):
    try:
        if math.isnan(x):
            return 0
    except TypeError:     
        return x.get('no_balls', 0)
    
def byes(x):
    try:
        if math.isnan(x):
            return 0
    except TypeError:     
        return x.get('byes', 0)
    
def legbyes(x):
    try:
        if math.isnan(x):
            return 0
    except TypeError:     
        return x.get('legbyes', 0)
```

```python
def create_match_dataframe(address):
    with open(address, 'r') as f:
        df = json.loads(f.read())
    match_df = pd.DataFrame({})
    for inning in range(2):
        df_1 = inning_df(df['innings'][inning]['overs'])
        df_2 = pd.DataFrame({'over': []})
        for i in range(20):
            temp1 = df_2.shape[0]
            df_2 = pd.concat([df_2, over_df(df_1.deliveries[i])], axis =0, ignore_index=True)
            temp2 = df_2.shape[0]
            df_2.loc[temp1:temp2, 'over'] = i+1
        match_df = pd.concat([match_df, df_2], axis = 0)
    match_df.reset_index(inplace = True)
    match_df['batter_runs'] = match_df.runs.apply(lambda x: dict(x)['batter'])
    match_df['extras_runs'] = match_df.runs.apply(lambda x: dict(x)['extras'])
    match_df['total_runs'] = match_df.runs.apply(lambda x: dict(x)['total'])
    match_df['player_out'] = match_df.wickets.apply(player_out)
    match_df['wicket_type'] = match_df.wickets.apply(wicket_type)
    match_df['wides'] = match_df.extras.apply(wides)
    match_df['legbyes'] = match_df.extras.apply(legbyes)
    match_df['byes'] = match_df.extras.apply(byes)
    match_df['no_ball'] = match_df.extras.apply(no_ball)
    match_df['fielders_involved_in_wicket'] = match_df.wickets.apply(fielders)
    return match_df
```

```python
match_df = create_match_dataframe('recently_played_30_male_json/1348650.json')
```
